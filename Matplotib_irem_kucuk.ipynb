```python
import matplotlib.pyplot as plt
#kütüphanemizi ekledik
```

```python
# Temel çizim örneği
x = [1, 2, 3, 4]
y = [15, 20, 25, 30]
plt.plot(x, y)
plt.show()# bu kısım önemli plot.show() ile görseleştirme işlemi yapabiliriz
```

```python
plt.plot(x, y, label="Doğru 1")
plt.plot(x, [i*2 for i in y], label="Doğru 2")#birden fazla doğruyu list copmheransion ile yapabiliriz
plt.legend()
plt.show()
```

```python
plt.plot(x, y, color="red", linestyle="--", marker="o", markersize=10, label="Özelleştirilmiş Çizgi")
plt.legend()
plt.show()
#Stil ve Özelleştirme kullanabilriz
```

```python
#Başlık ve Eksen Etiketleri
plt.plot(x, y)
plt.title("Grafik Başlığı")
plt.xlabel("X Ekseni Etiketi")
plt.ylabel("Y Ekseni Etiketi")
plt.show()
```

```python
plt.style.use('ggplot')#burada grafik use('ggplot') stilini  kullanadık
plt.plot(x, y)
plt.show()
```

```python
#Çubuk grafiği olarak kulanılır
plt.bar(x, y)
plt.show()
```

```python
#Histogramlar
data = [2, 2, 5, 3, 3, 3, 4, 8,6 , 4]
plt.hist(data, bins=4)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.arange(1,6)
b=np.arange(2,11,2)

fig=plt.figure() #Bir figür oluşturur
#axses=fig.add_axes([0.1,0.2,0.4,0.6]) =uzaklıkları [x deki çerçeve uzaklık,y deki çerçeve uzaklık,x uzaklığı,y uzaklığı]
axses=fig.add_axes([0.1,0.2,0.4,0.6])# Oluşturulan figür üzerinde alt bir eksen oluşturur. Koordinatlar [0.1, 0.2] ile başlar ve genişlik-yükseklik olarak [0.4, 0.6] kullanır.
axses1=fig.add_axes([0.4,0.6,0.8,0.10])#İkinci bir alt eksen oluşturur. Koordinatlar ve boyutlar aynıdır.
axses.plot(y,x)
axses.set_xlabel("outher x")#x eksenine başlık ile  isimlendirebiliriz
axses.set_ylabel("outher y")#y eksenine başlık ile  isimlendirebiliriz
axses.set_title("outher garph0 ")#Ana başlık ile  isimlendirebiliriz

axses1.plot(x,y)
#set_ylabellabel isim olarak verdik
axses1.set_xlabel("outher x")
axses1.set_ylabel("outher y")
axses1.set_title("outher garph1 ")
plt.show()
```

```python
#Bu satır, fig ve axes adlı iki değişken oluşturur. fig tüm figürü (yani grafik alanını) temsil ederken, axes ise alt grafiklerin (subplots) bulunduğu bir dizi veya listeyi temsil eder.
#Bu durumda, nrows=2 ve ncols=1 parametreleriyle 2 satırlık ve 1 sütunluk bir alt grafik ızgarası oluşturulur. Yani, üst üste iki grafik çizilecektir.
fig ,axes = plt.subplots(nrows=2,ncols=1)

print(axes[0])
print(axes[1])
print(type(axes))
plt.tight_layout()#Bu satır, grafiklerin birbirine girmesini önlemek ve daha düzenli görünmelerini sağlamak için otomatik olarak boşlukları ayarlar.
```

```python
#figürün düzlemin boyutunu verdik figsize
fig = plt.figure(figsize=(6,4))
#burada x ve y nin boyutunu verir
axes =fig.add_axes([0,0,1,1])#Bu satır, figürün içine bir eksen (axes) ekler. add_axes([left, bottom, width, height]) fonksiyonu, eksenlerin figür içindeki konumunu belirler.

axes.plot(x,x**5,color="red",label="kırmızı")
axes.plot(x,x**2,color="blue",label="mavi")
axes.plot(x,y,color="yellow",label="sarı")
#burda isimlerin gözükmesini istiyorsak bunu yazmamız gerekir
axes.legend()#Bu satır, grafiğe bir açıklama (legend) ekler. legend() fonksiyonu, yukarıda belirtilen etiketlerin (label parametreleri) grafikte görünmesini sağlar. Bu şekilde, hangi renkli çizginin hangi veriyi temsil ettiğini kolayca anlayabilirsiniz.
```

```python
fig = plt.figure(figsize=(6,4))

axes =fig.add_axes([0,0,1,1])

axes.plot(x,x**5,color="red",label="kırmızı",lw=3,linestyle=":",marker="o"
          ,markersize=20,markerfacecolor="black"
        )
"""
Bu satır, eksenler üzerine x ve x**5 verilerini çizer ve çeşitli stil özelliklerini uygular:

color="red": Çizginin rengi kırmızı olacaktır.
label="kırmızı": Çizgi için "kırmızı" etiketi belirlenir (bu, legend için kullanılır).
lw=3: Çizgi kalınlığı (line width) 3 olacak şekilde ayarlanır.
linestyle=":": Çizgi stili noktalı (dotted) olacaktır.
marker="o": Çizgi üzerindeki veri noktaları daire (o) şeklinde işaretlenecektir.
markersize=20: Veri noktalarının boyutu 20 olacaktır.
markerfacecolor="black": Veri noktalarının iç rengi siyah olacaktır.
Özetle, bu kod bir figür oluşturur, bu figüre bir eksen ekler ve bu eksen üzerinde x ve x**5 verilerini kırmızı renkte, kalın, noktalı çizgi stili ile ve daire şeklinde, büyük siyah işaretçilerle çizer. Bu stil özellikleri grafiği daha anlaşılır ve görsel olarak çekici hale getirir.
"""
```

```python
fig = plt.figure(figsize=(6,4))
#buda x ve y nin boyutunu verir
axes =fig.add_axes([0,0,1,1])

axes.plot(x,x**5,color="red",label="kırmızı",lw=3,linestyle=":",marker="o"
          ,markersize=20,markerfacecolor="black"
         , markeredgecolor="yellow",markeredgewidth=3)

axes.set_xlim(0,10)
axes.set_ylim(0,20)
"""
Bu satırlar, eksenlerin görünür aralığını ayarlar:

set_xlim(0, 10): X ekseninin aralığını 0 ile 10 arasında ayarlar.
set_ylim(0, 20): Y ekseninin aralığını 0 ile 20 arasında ayarlar.
Özetle, bu kod, belirli stil ve eksen ayarları ile bir çizgi grafiği oluşturur. Çizgi kırmızı renkte, kalın, noktalı çizgi stili ile ve büyük siyah iç, sarı kenar ve kalın kenar çizgili daire işaretçilerle çizilir. Ayrıca, eksenlerin görünür aralıkları X ekseni için 0-10 ve Y ekseni için 0-20 olarak belirlenmiştir.
"""
```
